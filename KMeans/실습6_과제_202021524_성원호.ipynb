# wine 데이터셋을 활용하여 K-means를 수행하시오.
##### "alcohol", "total_phenols" feature 사용

```python
from sklearn.datasets import load_wine
import seaborn as sns
import matplotlib.pyplot as plt

dataset = load_wine(as_frame=True)

df = dataset.frame
df["class"] = dataset.target

df[["alcohol","total_phenols"]]
```

```python
sns.scatterplot(data=df, x='alcohol', y='total_phenols', hue='class')
```

### 1. Elbow point를 시각화하세요.


```python
from sklearn.cluster import KMeans
```

```python
X = df[['alcohol', 'total_phenols']]
```

```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
y_pred
```

```python
kmeans.labels_
```

```python
y_pred is kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]
inertias
```

```python
plt.figure(figsize=(8,3.5))
plt.plot(range(1,10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.annotate('Elbow',
             xy=(4, inertias[3]),
             xytext=(0.55, 0.55),
             textcoords= 'figure fraction',
             fontsize=16,
             arrowprops=dict(facecolor='black', shrink=0.1)
            )
plt.axis([1,8.5,0,300])
plt.show()
```

### 3. 문제 1, 2의 결과를 토대로 적절한 k를 제시하고 그 이유에 대해서 설명하세요.

#### 문제 1에서의 엘보우 그래프를 통해 확인해보면 k = 4에서 SSE 값이 급격히 감소하고, 이후 감소폭이 완만해지는 것을 볼 수 있음. 따라서, k = 4일 때 최적의 클러스터라고 판단된다.
#### 문제 2는 각 클러스터 수 k에 대해 실루엣 계수를 시각화한 것으로, 실루엣 점수가 가장 높은 k를 선택하는 것이 최적 클러스터이므로, k = 2, 3일 때 가장 높은 실루엣 점수를 가지는 것으로 판단되어 k = 2,3일 때 최적의 클러스터 수로 선택될 수 있다.
#### 엘보우 그래프에서는 k=4 가 최적의 클러스터 수로 선택 되었으나, 실루엣 스코어 그래프에서 k=2,3의 스코어가 동일하고 가장 최적 클러스터 수로 판단되었다. 실루엣 스코어는 각 클러스터의 일관성을 측정하는 반면, 엘보우 메소드는 클러스터링 모델의 전체적 구조를 고려하므로, 사용 목적에 따라 적절한 최적 클러스터를 선택하여 사용하면 될 것이다.
#### 예를 들어 데이터의 자연적 구조를 파악하고자 할 때는 실루엣 스코어를, 클러스터링하여 분석 가능한 집단으로 그룹화하려는 경우 엘보우 메소드를 사용하면 된다.
